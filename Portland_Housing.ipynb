```python
import numpy as np
import pandas as pd
import seaborn as sns

data = pd.read_csv("portland_housing.csv")
df = data.copy()
df.head()
```

```python
drops=[]
for i in (df.corr()["price"].sort_values(ascending=False).index[20:]):
  drops.append(i)
drops
```

```python
df.drop(columns=drops,axis=1,inplace=True)
```

```python
df
```

```python
drops1=[]
for i in (df.dtypes == object).sort_values(ascending=True).index[15:] :
  drops1.append(i)
drops1
```

```python
df.drop(columns=drops1,axis=1,inplace=True)
```

```python
df.dropna(how='all',axis=1)
```

```python
drops3 = ["priceHistory/3/price","priceHistory/5/price","priceHistory/6/price","priceHistory/8/price","priceHistory/9/price"]
df.drop(columns=drops3,axis=1,inplace=True)
```

```python
df
```

```python
for i in df.isnull().sum().index:
  df[i].fillna(df[i].median(),inplace=True)
df
```

```python
df_price  = df['price']
Q1 = df_price.quantile(0.25)
Q3 = df_price.quantile(0.75)
IQR = Q3-Q1                        #ınter quartile
print(f"Q1 = {Q1}, Q3 = {Q3}, IQR= {IQR}")
alt_sinir = Q1-1.5*IQR
ust_sinir = Q3+1.5*IQR
print(f"alt sınır = {alt_sinir}, üst sınır = {ust_sinir}")
```

```python
aykiri_tf_ust  = df_price > ust_sinir
df_price[aykiri_tf_ust]
```

```python
df_price[aykiri_tf_ust] = ust_sinir
df_price[aykiri_tf_ust]
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_squared_error , r2_score
from sklearn.ensemble import RandomForestRegressor
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
y = df["price"]
X = df.drop(["price"],axis=1).astype("float64")
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
rf_model=RandomForestRegressor(random_state=42).fit(X_train,y_train)
y_pred = rf_model.predict(X_test)
r2_score(y_test,y_pred)
```
